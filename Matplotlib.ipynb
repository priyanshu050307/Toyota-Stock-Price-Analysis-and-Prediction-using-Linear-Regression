```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Matplotlib Basics

x = np.linspace(0,5,11)
y= x **2
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("Chacha")
plt.ylabel("Papa")
```

```python
plt.subplot(2,2,1)
plt.plot(x**2,y**9)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x*2,y**9)
```

#Object Oriented Way

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.2,0.5,0.5])
axis2.plot(x,y)
```
