```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv('final.csv')

print(df.head())
print(df.describe())
print(df.shape)
print(df.info())
print(df.isnull().sum())

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date')

df['Open_prev'] = df['Open'].shift(1)
df['Volume_prev'] = df['Volume'].shift(1)
df['RSI_prev'] = df['RSI'].shift(1)
df['MACD_prev'] = df['MACD'].shift(1)
df['Volatility_10_prev'] = df['Volatility_10'].shift(1)

df = df.dropna()

X = df[['Open_prev','Volume_prev','RSI_prev','MACD_prev','Volatility_10_prev']]
y = df['Close']

split = int(len(df)*0.8)
X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("MSE:", mean_squared_error(y_test,y_pred))
print("R2:",r2_score(y_test,y_pred))
coeff_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})

print(coeff_df)

last_row = df.iloc[-1]
input_data = [[
    last_row['Open'],
    last_row['Volume'],
    last_row['RSI'],
    last_row['MACD'],
    last_row['Volatility_10']
]]

predicted_price = model.predict(input_data)
print("Predicted Next Day Close:",predicted_price[0])


corr = df.corr(numeric_only=True)

plt.figure(figsize=(14, 12))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

#Objective1 - How does Toyota stock price evolve over time? Are there clear upward or downward trends?
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Close'])
plt.title("Toyota Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()

#Objective2 - Does higher trading volume lead to significant price changes?
plt.figure(figsize=(8,5))
plt.scatter(df['Volume'], df['Return'])
plt.title("Volume vs Return")
plt.xlabel("Volume")
plt.ylabel("Return")
plt.show()

```
